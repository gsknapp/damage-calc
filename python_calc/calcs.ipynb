## Ditto Testing

```python
ditto = pokemon.Pokemon(
    name="Ditto",
    level=100,
    gen=9,
    item="Choice Scarf",
    nature="Bold",
    evs={"hp" : 252, "spd" : 252}
    )
```

```python
chansey = pokemon.Pokemon(
    name="Chansey",
    item="Eviolite",
    nature="Calm",
    evs={"hp" : 252, "spd" : 252},
    gen=9)
```

```python
flip = move.Move(name="Flip Turn", gen=9)
```

```python
curr_field = field.Field(weather="Snow",terrain="Electric",defenderSide=Side(isReflect=True))
```

```python
calc.calculate(gen=9,attacker=ditto,defender=chansey,move=flip,field=curr_field)
```

```python
calc.calculate(gen=9,attacker=chansey,defender=chansey,move=flip,field=curr_field)
```

```python
new_ditto,chansey = calc.ditto_transform(ditto,chansey)
calc.calculate(gen=9,attacker=new_ditto,defender=chansey,move=flip,field=curr_field)
```

```python
chansey,new_ditto = calc.ditto_transform(chansey,ditto)
calc.calculate(gen=9,attacker=chansey,defender=new_ditto,move=flip,field=curr_field)
```
